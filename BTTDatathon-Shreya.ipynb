```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_dir1 = 'C:/Users/shrey/Downloads/Rockstar Games BTT Datathon Dataset'
file_name1 = 'player_statistics.csv'
file_path1 = os.path.join(base_dir1, file_name1)
df1 = pd.read_csv(file_path1)
base_dir2 = 'C:/Users/shrey/Downloads/Rockstar Games BTT Datathon Dataset'
file_name2 = 'item_spend.csv'
file_path2 = os.path.join(base_dir2, file_name2)
df2 = pd.read_csv(file_path2)
base_dir3 = 'C:/Users/shrey/Downloads/Rockstar Games BTT Datathon Dataset'
file_name3 = 'player_activity.csv'
file_path3 = os.path.join(base_dir3, file_name3)
df3 = pd.read_csv(file_path3)
merged_df_1_2 = pd.merge(df1, df2, on=['account_id', 'platform_id', 'occur_date'])
merged_df = pd.merge(merged_df_1_2, df3, on=['account_id', 'platform_id', 'occur_date'])
```

```python
merged_df.head()
```

```python
plt.scatter(merged_df['pvc_balance'], merged_df['evc_balance'])
plt.title('PVC Balance vs EVC Balance')
plt.xlabel('PVC Balance')
plt.ylabel('EVC Balance')
```

```python
plt.scatter(merged_df['time_spent'], merged_df['money_earned'])
plt.title('Money Earned vs Time Spent')
plt.xlabel('Time Spent')
plt.ylabel('Money Earned')
```

```python
merged_df['deaths'] = merged_df['deaths'].replace(0, 1)
merged_df['kd ratio'] = merged_df['kills']/merged_df['deaths']
```

```python
sns.barplot(x='platform_id', y='kd ratio', hue='platform_id', data=merged_df)
```
