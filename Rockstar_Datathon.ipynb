```python
full_df['K/D'] = np.divide(full_df['kills'], full_df['deaths'].replace(0, np.nan))
full_df['K/D'].fillna(0, inplace=True)
```

```python
full_df['evc_balance'].max()
```

```python
full_df['K/D']=np.multiply(full_df['K/D'],full_df['time_spent'])
full_df=full_df.rename(columns={'K/D':'Sentiment'})
```

```python
features = full_df[['platform_id']]
target = full_df['Sentiment']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
features_encoded = pd.get_dummies(full_df[['platform_id']], drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
full_df['predicted_sentiment'] = rf_model.predict(features_encoded)
ranked_players = full_df.sort_values(by='predicted_sentiment', ascending=False)
top_10_players = ranked_players.head(10)
print(top_10_players)

bottom_10_players = ranked_players.tail(10)
print(bottom_10_players)
```
